```python
df_non_covid_experiment = df_non_covid[df_non_covid.covid_status!='no_resp_illness_exposed']
np.unique(df_non_covid_experiment.covid_status)
```

```python
symptoms = ['cold', 'cough', 'fever', 'diarrhoea', 'st', 'loss_of_smell', 'mp', 'ftg', 'bd']
for symp in symptoms:
    df_temp = df_non_covid_experiment.fillna(-1)
    print(np.unique(df_temp[symp]))
```

```python
df_resp_ill = df_non_covid_experiment[(df_non_covid_experiment.covid_status=='resp_illness_not_identified')]
df_healthy_symp = df_non_covid_experiment[~(df_non_covid_experiment.covid_status=='resp_illness_not_identified')]
df_symp = df_healthy_symp.loc[(df_healthy_symp.cold==True) | (df_healthy_symp.cough==True) | (df_healthy_symp.fever==True) | (df_healthy_symp.diarrhoea==True) | (df_healthy_symp.st==True) | (df_healthy_symp.loss_of_smell==True) | (df_healthy_symp.mp==True) | (df_healthy_symp.ftg==True) | (df_healthy_symp.bd==True)]
df_healthy = df_healthy_symp.loc[~(df_healthy_symp.id.isin(df_symp.id.values))]
```
